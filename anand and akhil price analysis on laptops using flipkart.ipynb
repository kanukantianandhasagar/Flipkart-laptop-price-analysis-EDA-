```python
import pandas as pd
```

```python
ratings = []
prices = []
ROM = []
RAM = []
brands = []
Inches = []
processor = []
for i in range(1,46):
    url =f'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page={i}'
    
    
    print(url)
    
    #requests.get(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    
    
    
    for i in soup.find_all('div', class_='_2kHMtA'):
    #print(i)
        a = i.find('div', class_='_3LWZlK')
        if a:
            
            ratings.append(a.text)
        else:
            ratings.append(np.nan)


    ####price##
        price = i.find('div', class_='_30jeq3 _1_WHN1')
        #print(price.text)
        if price:
            prices.append(price.text)
        else:
            prices.append(np.nan)

    ### rom##
        rom =i.find('div', class_='_4rR01T')
        regex = re.findall(r'\b(\d+\s?(?:TB|GB)\s(?:SSD|HDD))\b', rom.text)
        if regex:
            ROM.append(regex[0])
        else:
            ROM.append(np.nan)
            
            
   ### ram ##
        ram =i.find('div', class_='_4rR01T')
        regex = re.findall(r'\d{1,3}\sGB', ram.text)
        if regex:
            RAM.append(regex[0])
        else:
            RAM.append(np.nan)
            
    ##brand
        b  = i.find('div', class_='_4rR01T')
        #brand = i.text
        regex = re.findall(r'^.\w+', b.text)
        if regex:
            brands.append(regex[0])
        else:
            brands.append(np.nan)
        
        
        ###inches
        inches = i.find('ul', class_="_1xgFaf")
        regex = re.findall(r'\(([^)]*[Ii]nch(?! Gen)[^)]*)\)',inches.text)
        #print(regex) 
        if regex:
            Inches.append(regex[0])
        else:
            Inches.append(np.nan)
        
        
        pro = i.find('ul', class_='_1xgFaf')
    #print(i.text)
    
        regex = re.findall(r'^([^(\n]+)(?=\sProcessor)', pro.text)
    
        if regex:
            processor.append(regex[0])
        else:
            processor.append(np.nan)
```

```python
len(brands)
```

```python
df = pd.DataFrame({'Brand':brands,'Prices':prices,'ROM':ROM,'RAM':RAM,'Ratings':ratings,'Inches':Inches,'Processor':processor})
```

```python
df['RAM'] = df['RAM'].astype(str).str.extract('(\d+)', expand=False)
df['RAM'] = pd.to_numeric(df['RAM'], errors='coerce')
```

```python
df['ROM'] = df['ROM'].astype(str).str.extract('(\d+)', expand=False)
df['ROM'] = pd.to_numeric(df['ROM'], errors='coerce')
```

```python
df['Inches'] = df['Inches'].astype(str).str.extract('(\d+)', expand=False)
df['Inches'] = pd.to_numeric(df['Inches'], errors='coerce')
```

```python
import numpy as np
valid_processor_patterns = ["^Intel", "^AMD", "^Apple", "^Media"]
df['Processor'] = np.where(df['Processor'].str.contains('|'.join(valid_processor_patterns), case=False, na=False), df['Processor'], np.nan)
```
